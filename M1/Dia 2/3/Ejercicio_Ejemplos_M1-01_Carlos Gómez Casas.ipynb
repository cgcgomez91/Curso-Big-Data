#Usar nombres de variables significativos; para usarlos tanto tu como otras personas

```python
#Pone un nombre con significado a la variable
total_min= 43 + 42 + 57 #duración total de varios álbumes en minutos
print(total_min) 
```

```python
total_hours = total_min / 60 #duración total de varios albumes en horas
print(total_hours)
```

```python
total_hours = (43 + 42 + 57) / 60 #horas totales en una sola expresión
print(total_hours)
```
